### Проанализировав датафреймы видим что в них разное количество столбцов, пропусков в данных нет.  Для того чтобы сделать один большой датафрейм нужно привести данные к общему виду.  Количество стобцов которые есть во всех датафреймах - 7

```python
happiness_2015['Year'] = 2015
happiness_2016['Year'] = 2016
happiness_2017['Year'] = 2017
happiness_2018['Year'] = 2018
happiness_2019['Year'] = 2019
```

```python
happiness_2015_grouped = happiness_2015.loc[:, 
    ['Happiness Rank', 'Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)',
    'Freedom', 'Generosity', 'Trust (Government Corruption)', 'Year']]
happiness_2015_grouped.head()                     
```

```python
happiness_2016_grouped = happiness_2016.loc[:, [
    'Happiness Rank', 'Country', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)',
    'Freedom', 'Generosity', 'Trust (Government Corruption)', 'Year']]
happiness_2016_grouped.head()
```

```python
happiness_2017_grouped = happiness_2017.loc[:, 
                ['Happiness.Rank', 'Country', 'Happiness.Score', 'Economy..GDP.per.Capita.', 
                 'Health..Life.Expectancy.', 'Freedom', 'Generosity', 'Trust..Government.Corruption.', 'Year']]
happiness_2017_grouped.head()
```

```python
happiness_2018_grouped = happiness_2018.loc[:, 
    ['Overall rank', 'Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy',
     'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Year']]
happiness_2018_grouped.head()
```

```python
happiness_2019_grouped = happiness_2019.loc[:, 
    ['Overall rank', 'Country or region', 'Score', 'GDP per capita', 'Healthy life expectancy',
     'Freedom to make life choices', 'Generosity', 'Perceptions of corruption', 'Year']]
happiness_2019_grouped.head()
```

```python
def stndr_colums(df_of_year):
    """Принимаем на вход датафрейм и приводим названия его колонок к общему виду"""
    
    dict_of_colums = {}
    list_of_colums = ['happiness_rank', 'country', 'happiness_score', 'GDP_per_capita',
                      'healthy_life_expectancy', 'freedom', 'generosity', 'perceptions_of_corruption', 'year']
    for i, colume in enumerate(df_of_year.columns):
        dict_of_colums[colume] = list_of_colums[i]
    
    df_of_year.rename(columns=dict_of_colums, inplace=True)
```

```python
stndr_colums(happiness_2015_grouped)
happiness_2015_grouped.head()
```

```python
stndr_colums(happiness_2016_grouped)
stndr_colums(happiness_2017_grouped)
stndr_colums(happiness_2018_grouped)
stndr_colums(happiness_2019_grouped)
```

```python
db_happiness = pd.concat([happiness_2015_grouped, happiness_2016_grouped, happiness_2017_grouped,
                          happiness_2018_grouped, happiness_2019_grouped], ignore_index=True, copy=True)
```

```python
db_happiness_all = db_happiness
db_happiness_all
```

Найдем самую среднюю страну

```python
def mean_count(df_of_year):
    """Принимаем датафрейм счастья стран за год и находим название самых средних по счастью стран"""
    
    median_rank = df_of_year['happiness_rank'].median().round(0)
    return df_of_year.loc[df_of_year['happiness_rank'].isin([median_rank, median_rank - 1, median_rank + 1])]
    
```

```python
mean_count(happiness_2015_grouped)
```

```python
mean_count(happiness_2016_grouped)
```

```python
mean_count(happiness_2017_grouped)
```

```python
mean_count(happiness_2018_grouped)
```

```python
mean_count(happiness_2019_grouped)
```
