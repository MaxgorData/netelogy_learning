```python
log.loc[
        ((log.traffic_source == 'paid') | (log.traffic_source == 'email'))
        &
        (log.region != 'Russia')
        , 'source_type'] = 'other'
```

## Задание 3
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
rate = pd.read_csv('ml-latest-small/ratings.csv')
rate.head()
```

```python
rates_of_users = rate.groupby('userId').count()
```

```python
rates_of_users.head()
```

```python
rates_of_users = rates_of_users[rates_of_users['movieId'] >= 100].reset_index()
```

```python
rates_of_users
```

```python
def user_lifetime(userId):
    """Высчитываем время жизни пользователя"""
    
    return rate[rate['userId'] == userId].timestamp.max() - rate[rate['userId'] == userId].timestamp.min()
```

```python
rates_of_users['lifetime'] = rates_of_users['userId'].apply(user_lifetime)
```

```python
rates_of_users[['userId', 'lifetime']]
```
