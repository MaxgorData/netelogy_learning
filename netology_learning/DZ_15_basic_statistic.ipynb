```python
f = open('horse_data.names', 'r')
horse_data = f.readlines()
f.close()
horse_data
```

```python
def colums_for_df(horse_data):
    """Принимает текст и выводит пронумерованный словарь названий колонок"""
    re_horse = []
    horse_names_colums = {}
    horse_colums = {}
    
    for line in horse_data:
        find = re.findall(r'\d\:\s+(\w+)', line)
        if find:
            re_horse.append(re.findall(r'\d+\:\s+.+', line))
    
    for string in re_horse:
        ind = re.search(r'\d+', string[0])
        colum = re.findall(r'\d\:\s+(.+)', string[0])[0]
        for i in range(1,29):
            if i == int(ind[0]):
                horse_names_colums[i] = colum
    
    for i in range(1,29):
        if i in list(horse_names_colums.keys()):
            horse_colums[i] = horse_names_colums[i]
        else:
            horse_colums[i] = i
    
    return horse_colums
```

```python
horse_dirt = colums_for_df(horse_data)
print(list(horse_dirt.values()))
```

Крайне непонятно что делать с данными без цели, поэтому исходил из самых явных целей и в частности исхода пациента
